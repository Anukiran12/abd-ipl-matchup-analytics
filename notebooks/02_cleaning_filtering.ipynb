```python
# now we can define our targets 

batter="AB de Villiers"

bowlers=[
    "JJ Bumrah",
    "DJ Bravo",
    "DW Steyn",
    "SL Malinga",
    "R Ashwin",
    "RA Jadeja",
    "SP Narine"

]
```

```python
filtered_df=df[
    (df['batter']==batter) &
    (df['bowler'].isin(bowlers))
]

filtered_df.head()
```
